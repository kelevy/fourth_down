```python
go_rate = (
    fourth
    .groupby(pd.cut(fourth["ydstogo"], bins=[0,1,2,3,5,8,12,20,100]))["decision"]
    .apply(lambda s: (s == "go_for_it").mean())
)

plt.figure(figsize=(7,4))
plt.plot(go_rate.index.astype(str), go_rate.values, marker="o")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Go-for-it rate")
plt.xlabel("Yards to go")
plt.title("4th-down go-for-it rate by yards to go")
plt.tight_layout()
plt.show()
```
