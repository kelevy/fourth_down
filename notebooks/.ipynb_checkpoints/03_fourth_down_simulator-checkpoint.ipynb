```python
import sys
from pathlib import Path

PROJECT_ROOT = Path("..").resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import nfl_data_py as nfl
```

```python
pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 140)

SEASONS = [2022, 2023, 2024]
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
```

```python


fourth_path = PROCESSED_DIR / "fourth_down.parquet"
pbp_plays_path = PROCESSED_DIR / "pbp_plays.parquet"


fourth = pd.read_parquet(fourth_path)
print("Loaded fourth_down:", fourth.shape)

if pbp_plays_path.exists():
    pbp = pd.read_parquet(pbp_plays_path)
    print("Loaded pbp_plays:", pbp.shape)
else:
    print("pbp_plays.parquet not found â†’ importing pbp directly (cache disabled).")
    pbp = nfl.import_pbp_data(SEASONS, downcast=True, cache=False)
    pbp = pbp[pbp["play_type"].notna()].copy()
    pbp = pbp[pbp["down"].isin([1,2,3,4])].copy()
    print("Imported plays:", pbp.shape)

fourth.head()
```
