```python
import sys
from pathlib import Path

# Make ../src importable when running from notebooks/
PROJECT_ROOT = Path("..").resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nfl_data_py as nfl
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss, brier_score_loss, roc_auc_score
import json


pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 140)
```

```python
SEASONS = [2022, 2023, 2024]
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
```

```python
pbp = nfl.import_pbp_data(SEASONS, downcast=True, cache=False)
print("Loaded pbp:", pbp.shape)

# Keep actual plays (drop markers/timeouts/etc.)
pbp = pbp[pbp["play_type"].notna()].copy()
pbp = pbp[pbp["down"].isin([1,2,3,4])].copy()
print("Filtered plays:", pbp.shape)

pbp.head()
```

```python
games = (
    pbp.groupby("game_id", as_index=False)
       .agg(
           season=("season", "first"),
           week=("week", "first"),
           game_date=("game_date", "first") if "game_date" in pbp.columns else ("season","first"),
           home_team=("home_team", "first"),
           away_team=("away_team", "first"),
           home_score_final=("home_score", "max"),
           away_score_final=("away_score", "max"),
       )
)

games["home_win"] = (games["home_score_final"] > games["away_score_final"]).astype(int)

pbp = pbp.merge(games[["game_id", "home_win"]], on="game_id", how="left")
pbp[["game_id","home_team","away_team","home_win"]].head()
```

```python
# Core numeric fields weâ€™ll use
for c in ["down","ydstogo","yardline_100","qtr","game_seconds_remaining","half_seconds_remaining",
          "home_score","away_score","home_timeouts_remaining","away_timeouts_remaining"]:
    if c in pbp.columns:
        pbp[c] = pd.to_numeric(pbp[c], errors="coerce")

# Possession & score from home perspective
pbp["is_home_possession"] = (pbp["posteam"] == pbp["home_team"]).astype(int)
pbp["score_diff_home"] = pbp["home_score"].fillna(0) - pbp["away_score"].fillna(0)

# Time remaining in game
pbp["time_remaining"] = pbp["game_seconds_remaining"].fillna(pbp.get("half_seconds_remaining", 0)).clip(lower=0)

# Timeouts (optional but useful; fill missing with 3)
pbp["home_to"] = pbp.get("home_timeouts_remaining", pd.Series(np.nan, index=pbp.index)).fillna(3).clip(0, 3)
pbp["away_to"] = pbp.get("away_timeouts_remaining", pd.Series(np.nan, index=pbp.index)).fillna(3).clip(0, 3)

# Clip a couple of fields to keep things sane
pbp["ydstogo"] = pbp["ydstogo"].clip(0, 50)
pbp["yardline_100"] = pbp["yardline_100"].clip(1, 99)
pbp["qtr"] = pbp["qtr"].clip(1, 5)

# Canonical modeling dataframe
df = pbp.dropna(subset=["home_win", "down", "ydstogo", "yardline_100", "time_remaining"]).copy()
print("Model df:", df.shape)
df.head()
```

```python
NUM_FEATURES = [
    "down", "ydstogo", "yardline_100", "qtr", "time_remaining",
    "score_diff_home", "home_to", "away_to", "is_home_possession"
]

def make_model():
    num = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()),
    ])
    pre = ColumnTransformer([("num", num, NUM_FEATURES)], remainder="drop")
    clf = LogisticRegression(max_iter=300)
    return Pipeline([("pre", pre), ("clf", clf)])

def evaluate(y_true, y_prob):
    y_prob = np.clip(y_prob, 1e-6, 1-1e-6)
    return {
        "log_loss": float(log_loss(y_true, y_prob)),
        "brier": float(brier_score_loss(y_true, y_prob)),
        "auc": float(roc_auc_score(y_true, y_prob)),
    }
```

```python
rng = np.random.default_rng(7)
game_ids = df["game_id"].unique()
rng.shuffle(game_ids)

split = int(0.8 * len(game_ids))
train_games = set(game_ids[:split])
test_games  = set(game_ids[split:])

train = df[df["game_id"].isin(train_games)].copy()
test  = df[df["game_id"].isin(test_games)].copy()

print("Train games:", len(train_games), "Test games:", len(test_games))
print("Train plays:", len(train), "Test plays:", len(test))
```

```python
model = make_model()

X_train = train
y_train = train["home_win"].astype(int).values

X_test = test
y_test = test["home_win"].astype(int).values

model.fit(X_train, y_train)
p_test = model.predict_proba(X_test)[:, 1]

metrics = evaluate(y_test, p_test)
metrics
```

```python
def reliability_diagram(y_true, y_prob, n_bins=12):
    y_true = np.asarray(y_true)
    y_prob = np.asarray(y_prob)

    bins = np.linspace(0, 1, n_bins + 1)
    b = np.digitize(y_prob, bins) - 1
    b = np.clip(b, 0, n_bins - 1)

    xs, ys, counts = [], [], []
    for i in range(n_bins):
        m = (b == i)
        if m.sum() == 0:
            continue
        xs.append(y_prob[m].mean())
        ys.append(y_true[m].mean())
        counts.append(m.sum())

    plt.figure(figsize=(6,6))
    plt.plot([0,1],[0,1], "--")
    plt.plot(xs, ys, marker="o")
    plt.xlabel("Predicted P(home wins)")
    plt.ylabel("Empirical home win rate")
    plt.title("Reliability Diagram")
    plt.tight_layout()
    plt.show()

reliability_diagram(y_test, p_test, n_bins=12)
```

```python
tmp = test.copy()
tmp["p_home_win"] = p_test

late = tmp[tmp["time_remaining"] < 600].copy()  # last 10 minutes
bins = pd.cut(late["score_diff_home"], bins=np.arange(-28, 29, 7), include_lowest=True)
grp = late.groupby(bins)["p_home_win"].mean()

plt.figure(figsize=(8,4))
plt.plot(grp.index.astype(str), grp.values, marker="o")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Mean predicted P(home wins)")
plt.title("Sanity check: WP vs score differential (last 10 minutes)")
plt.tight_layout()
plt.show()
```

```python
(PROCESSED_DIR / "wp_model_metrics.json").write_text(json.dumps(metrics, indent=2), encoding="utf-8")

sample = test.sample(min(10000, len(test)), random_state=1).copy()
sample["p_home_win"] = model.predict_proba(sample)[:, 1]
sample[[
    "game_id","play_id","home_team","away_team","posteam",
    "down","ydstogo","yardline_100","qtr","time_remaining",
    "score_diff_home","p_home_win","home_win"
]].to_parquet(PROCESSED_DIR / "wp_predictions_sample.parquet", index=False)

print("Saved:")
print("-", PROCESSED_DIR / "wp_model_metrics.json")
print("-", PROCESSED_DIR / "wp_predictions_sample.parquet")
```
